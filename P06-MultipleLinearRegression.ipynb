```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/myanmards/resource_files/master/regression-one.csv')
df.head()
```

```python
df.rename(columns= {'Life expectancy': 'life-expect'}, inplace = True)
df.head()
```

```python
df = df.dropna()
df.shape
```

```python
df[df.isnull().any(axis=1)]
```

```python
X = df[['GDP', 'Alcohol', 'BMI', 'Schooling']]
Y = df['life-expect']
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X, Y)
```

lm.intercept_

```python
lm.coef_
```

```python
Y_hat = lm.predict(X)
```

```python
data = pd.DataFrame({'Actual':df['life-expect'], 'Predicted': Y_hat})
data.head(20)
```

```python
import seaborn as sns

axs = sns.distplot(df['life-expect'], hist=False, color="r", label="Actual Value")
sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values", ax=axs)

plt.title('Acutal vs Fitted Values for LE')
plt.xlabel('Age Range')
plt.show()
```
